# Interpolación en 2 puntos

```julia
L1(x, x1, x2) = (x - x2) / (x1 - x2)
```

```julia
L2(x, x1, x2) = L1(x, x2, x1)
```

```julia
x1, x2 = 1, 3
y1, y2 = 2, 7

L1(x1, x1, x2), L1(x2, x1, x2)
```

```julia
-0.0 == 0.0
```

```julia
-0.0 === 0.0
```

```julia
1 / Inf
```

```julia
-1 / Inf
```

```julia
L(x, x1, x2, y1, y2) = y1 * L1(x, x1, x2) +
                        y2 * L2(x, x1, x2)
```

```julia
xx = 0:0.1:4  # los puntos en donde quiero calcular la función interpoladora
yy = [ L(x, x1, x2, y1, y2) for x in xx ]
```

```julia
gr(size=(300, 300))
```

```julia
using Interact
```

```julia
r = -3:0.01:3
@manipulate for x1 in r, x2 in r, y1 in r, y2 in r
    xx = -4:0.1:4  # los puntos en donde quiero calcular la función interpoladora
    yy = [ L(x, x1, x2, y1, y2) for x in xx ]
    
    scatter([x1, x2], [y1, y2], m=:square, leg=:topleft, label="puntos")
    plot!(xx, yy, m=:o, ms=1, label="interpolación")  # ms es "markersize" (tamaño de marcadores)
    scatter!(xx, [0 for x in xx], ms=1)

    scatter!(xx, L1.(xx, x1, x2), label="L1")
    scatter!(xx, L2.(xx, x1, x2), label="L2")

    scatter!([x1], [0], ms=10, m=:x)
    scatter!([x2], [0], ms=10, m=:x, leg=false)
    
    ylims!(-10, 10)
end
```
